```python
# Exact code
from PIL import Image
import random
from numpy import array
from matplotlib import pyplot as plt
im_1 = Image.open(r"gry.jpg")
ar = array(im_1)

def GiveRand(st,r,Done):
    while True:
        i = random.randint(st,r-1)
        if i not in Done:
            Done.append(i)
            break
    return i

def crossover(ar):
    rL=len(ar)
    cL = len(ar[0])
    Done = []
    flag = True
#     t1= 0
#     t2= rL-1
    for i in range(0,rL,2):
        i1 = GiveRand(0,rL/2,Done)
        i2 = GiveRand(rL/2,rL,Done)
        D =[]
        s= int(cL/2)+int(cL/4)
        for c in range(s):
            ind = GiveRand(0,cL,D)
            ar[i1][ind],ar[i2][ind] = ar[i2][ind],ar[i1][ind]           
            
            #Mutation
          
            temp = random.randint(90,150) + ar[i1][ind]
            if(temp > 255):
                temp = 255
            ar[i1][ind] = temp 
            temp = random.randint(90,150) + ar[i2][ind]
            if(temp > 255):
                temp = 255
            ar[i2][ind] = temp 
            
def fitness(ar):
    lis=[]
    for i in range(0,len(ar)):
        full=len(ar[i])
        first,last=0,0
        sum=0
        if (full/2) == 0:    # checking if columns are even
            first,last=full/2,full/2
        else:
            first=int(full/2)
            last=int(full/2)
            last=last+1
            
            
            
        for j in range(0,first):
            sum=sum+ar[i][j]
            
        for j in range(0+first,last+first-1):
            sum=sum-ar[i][j]
        lis.append(sum)
    return lis

def sort(arr,fitness):
    for i in range(0,len(arr)):
        for j in range(0,len(fitness)):
            if fitness[i]<fitness[j]:
                fitness[i],fitness[j] = fitness[j],fitness[i]
                arr[i],arr[j] = arr[j],arr[i]    
        
print("after encryption : ")
#print(ar.shape)
fit = fitness(ar)
#print(fit)
#print(fit)
sort(ar,fit)
#print("sorted")
crossover(ar)

#imgplot = plt.imshow(ar, cmap='gray')
#imgplot
```
